##  1. Crea un DataFrame frutas que luzca así:

![Reto Uno]()

```python
import pandas as pd

datos = {
    'Granadilla': pd.Series([20]),
    'Tomates': pd.Series([50])
    }
frutas = pd.DataFrame(datos)
print(frutas)
```

Creamos un diccionario con dos claves: 'Granadilla' y 'Tomates', cada una con una serie de Pandas que tiene un solo valor. Luego, convertimos el diccionario en un DataFrame con pd.DataFrame(datos) y lo mostramos con print(frutas).

##  2. Crea un DataFrame ventas_frutas que coincida con el diagrama:

![reto dos]()

```python
import pandas as pd

datos = {
    'Granadilla': pd.Series([20, 49], index=["ventas 2021","ventas 2022"]),
    'Tomates': pd.Series([50, 100], index=["ventas 2021","ventas 2022"])
    }
ventas_frutas = pd.DataFrame(datos)
print(ventas_frutas)
```

Creamos un diccionario con dos claves: 'Granadilla' y 'Tomates', donde cada una tiene una serie de Pandas con datos de ventas para los años 2021 y 2022.

Usamos el parámetro index para asignar etiquetas a las filas. Luego, convertimos el diccionario en un DataFrame con pd.DataFrame(datos) y lo mostramos con print(ventas_frutas).

## 3. Crea una variable utensilios con una Serie que tenga el siguiente aspecto:

![reto tres]()

```python
import pandas as pd

utensilios = pd.Series(["3 unidades",
                        "2 unidades",
                        "4 unidades",
                        "5 unidades"], index=["Cuchara",
                                              "Tenedor",
                                              "Cuchillo",
                                              "Plato"], name="Cocina")
print(utensilios)
```

## 7. ¿Cuál es el precio promedio?

```python
precio_promedio = wine_review_api['price'].mean() 
print(f"El precio promedio es: {precio_promedio}")
```

## 8. ¿Cuál es el precio más alto pagado?

```python
precio_maximo = wine_review_api['price'].max()
print(f"El precio más alto pagado es: {precio_maximo}")
```

con el metodo max() encontramos el precio maximo dentro de ["price"]

## 10. Utiliza idxmax() para encontrar el índice del vino con el precio más alto y luego utiliza loc para obtener toda la información de ese vino específico.
## Salida esperada:

![reto diez]()

```python
indice_vino_caro = wine_review_api['price'].idxmax() 

indice_vino_caro_info = wine_review_api.loc[indice_vino_caro] 

indice_vino_caro_info
```

idxmax() nos da el índice donde el precio es máximo, y loc[] nos permite obtener toda la fila con la información del vino más caro.
