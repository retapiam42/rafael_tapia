## Broadcasting e indexado de Arrays

### 6. Crea un array de tama침o 3x1 y uno de 1x3, y s칰malos utilizando broadcasting para obtener un array de 3x3.

```python
import numpy as np

array_3x1 = np.array([[1], [2], [3]])

array_1x3 = np.array([[4, 5, 6]])

resultado = array_3x1 + array_1x3

print("Array 3x1:\n", array_3x1)
print("Array 1x3:\n", array_1x3)
print("Resultado de la suma (3x3):\n", resultado)
```

Primero, generamos un array de dimensiones 3x1 y otro de 1x3. Luego, aplicamos broadcasting para realizar la suma entre ambos, obteniendo un array resultante de 3x3, y finalmente mostramos el resultado.

### 12. Genera un gr치fico de dispersi칩n las variables 洧논 y 洧녽 = 洧멇롐뒳롐(洧논)+ ruido Gaussiano. Donde x es un array con n칰mereos entre -2洧랢 洧녽 2洧랢. Grafica tambi칠n los puntos 洧녽 = 洧멇롐뒳롐(洧논) en el mismo plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generar los valores de x
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

ruido = np.random.normal(0, 0.2, x.shape)

y_con_ruido = np.sin(x) + ruido

y_sin_ruido = np.sin(x)

plt.scatter(x, y_con_ruido, label='y = sin(x) + ruido gaussiano', color='red')
plt.plot(x, y_sin_ruido, label='y = sin(x)', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('gr치fico de dispersi칩n con ruido Gaussiano')
plt.legend()
plt.show()
```

Generamos un conjunto de valores x en el rango de 2 a 2 utilizando np.linspace(). Luego, a침adimos ruido gaussiano con np.random.normal(0, 0.2, x.shape) a la funci칩n seno (np.sin(x)).

Graficamos los puntos ruidosos con plt.scatter() en color rojo y la funci칩n seno original con plt.plot() en azul. Finalmente, agregamos etiquetas, t칤tulo y una leyenda antes de mostrar el gr치fico con plt.show().

### 13. Utiliza la funci칩n np.meshgrid para crear una cuadr칤cula y luego aplica la funci칩n z = np.cos(x) + np.sin(y) para generar y mostrar un gr치fico de contorno.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.linspace(-2 * np.pi, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)

Z = np.cos(X) + np.sin(Y)

plt.contour(X, Y, Z)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('grafico de Contorno de z = cos(x) + sin(y)')
plt.colorbar(label='Valor de z')
plt.show()
```

Generamos dos arreglos x y y en el rango de 2 a 2  con np.linspace(). Luego, creamos una malla 2D con np.meshgrid(x, y), y calculamos Z=cos(X)+sin(Y).

Graficamos las l칤neas de contorno con plt.contour(), a침adimos etiquetas a los ejes, un t칤tulo y una barra de color (plt.colorbar()) para indicar los valores de Z. Finalmente, mostramos el gr치fico con plt.show().

### 14. Crea un gr치fico de dispersi칩n con 1000 puntos aleatorios y utiliza la densidad de estos puntos para ajustar el color de cada punto.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.rand(1000)
y = np.random.rand(1000)

plt.figure(figsize=(10, 6))
sns.kdeplot(x, y, cmap='Blues', shade=True, thresh=0.05)
plt.scatter(x, y, c='blue', s=5, edgecolor='none')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('grafico de dispersi칩n con densidad de Puntos')
plt.show()
```

Generamos 1000 puntos aleatorios para x e y usando np.random.rand(). Luego, creamos un gr치fico de dispersi칩n con plt.scatter(), donde los puntos se representan en azul.

Adem치s, usamos sns.kdeplot() para a침adir un mapa de densidad con un degradado en tonos azules (cmap='Blues'), resaltando las 치reas con mayor concentraci칩n de puntos. Finalmente, configuramos etiquetas, t칤tulo y mostramos el gr치fico.

### 15. A partir de la misma funci칩n del ejercicio 12, genera un gr치fico de contorno lleno.

### 17. Genera dos sets de datos con distribuciones normales diferentes y mu칠stralos en el mismo histograma.

Luego, graficamos ambos en un histograma superpuesto, ajustando la transparencia (alpha=0.5) para visualizar mejor la superposici칩n. A침adimos etiquetas, una leyenda y mostramos el gr치fico con plt.show().

### 18. Experimenta con diferentes valores de bins (por ejemplo, 10, 30, 50) en un histograma y observa c칩mo cambia la representaci칩n.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

bins_values = [10, 30, 50]

plt.figure(figsize=(18, 6))

for i, bins in enumerate(bins_values, start=1):
    plt.subplot(1, 3, i)  
    plt.hist(data, bins=bins, edgecolor='black')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma con {bins} bins')

plt.tight_layout()  
plt.show()
```

Generamos un conjunto de 1000 datos aleatorios con distribuci칩n normal est치ndar usando np.random.randn().

Luego, creamos tres histogramas en una misma figura con diferentes cantidades de bins (10, 30 y 50), usando plt.subplot(1, 3, i) para organizarlos en una fila y tres columnas.

Cada histograma tiene bordes negros (edgecolor='black'), etiquetas y un t칤tulo indicando la cantidad de bins utilizada. Finalmente, usamos plt.tight_layout() para evitar que los gr치ficos se superpongan y mostramos la figura con plt.show().

### 19. A침ade una l칤nea vertical que indique la media de los datos en el histograma.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(1000)

media = np.mean(data)

bins_values = [10, 30, 50]

plt.figure(figsize=(18, 6))

for i, bins in enumerate(bins_values, start=1):
    plt.subplot(1, 3, i)  
    plt.hist(data, bins=bins, edgecolor='black')
    plt.axvline(media, color='red', linestyle='dashed', linewidth=1, label=f'Media: {media:.2f}')
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.title(f'Histograma con {bins} bins')
    plt.legend()

plt.tight_layout()  
plt.show()
```

Generamos un conjunto de 1000 datos aleatorios con distribuci칩n normal est치ndar usando np.random.randn(), y calculamos su media con np.mean().

Luego, creamos tres histogramas con diferentes cantidades de bins (10, 30 y 50), agregando una l칤nea vertical punteada en rojo (plt.axvline()) para indicar la media en cada gr치fico.

Cada histograma incluye etiquetas, una leyenda con el valor de la media y un dise침o optimizado con plt.tight_layout(). Finalmente, mostramos la figura con plt.show().
